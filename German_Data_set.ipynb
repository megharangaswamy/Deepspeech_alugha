```python
#1) separate xml file to different dataframes
df1_dev = pd.DataFrame(columns=['sentence','clean_sentence'])
inputdir_dev = '/home/megha/german dataset/german-speechdata-package-v2/dev'    # this only dev 
inputdir_train = '/home/megha/german dataset/german-speechdata-package-v2/train' # this only train 
inputdir_test = '/home/megha/german dataset/german-speechdata-package-v2/test' # this only test

df1_dev = pd.DataFrame(columns=['sentence','cleaned_sentence'])
df1_train = pd.DataFrame(columns=['sentence','cleaned_sentence'])
df1_test = pd.DataFrame(columns=['sentence','cleaned_sentence'])

df1_wav_dev = pd.DataFrame(columns=['transcript','wave_size'])
df1_wav_train = pd.DataFrame(columns=['transcript','wave_size'])
df1_wav_test = pd.DataFrame(columns=['transcript','wave_size'])

```

```python
#dev


for filename in os.listdir(inputdir_dev): 
    if filename.endswith(".xml"):
        #print(filename)     
        df2_dev = pd.DataFrame([filename])
        df1_dev = df1_dev.append(df2_dev, ignore_index = True)
    if filename.endswith(".wav"):
        #print(filename)     
        df2_wav_dev = pd.DataFrame([filename])
        df1_wav_dev = df1_wav_dev.append(df2_wav_dev, ignore_index = True)
        
df1_dev = df1_dev.rename(index=str, columns={0 : "xml_files"})
df1_wav_dev = df1_wav_dev.rename(index=str, columns={0 : "file_path"})


# updating file path column with full path
for i in range(0, len(df1_wav_dev)):
    file_name = df1_wav_dev['file_path'].iloc[i]
    file_name_full_path = '/home/megha/german dataset/german-speechdata-package-v2/dev/'+file_name
    df1_wav_dev['file_path'].iloc[i] = file_name_full_path
```

```python
#train


for filename in os.listdir(inputdir_train): 
    if filename.endswith(".xml"):
        #print(filename)     
        df2_train = pd.DataFrame([filename])
        df1_train = df1_train.append(df2_train, ignore_index = True)
    if filename.endswith(".wav"):
        #print(filename)     
        df2_wav_train = pd.DataFrame([filename])
        df1_wav_train = df1_wav_train.append(df2_wav_train, ignore_index = True)
        
df1_train = df1_train.rename(index=str, columns={0 : "xml_files"})
df1_wav_train = df1_wav_train.rename(index=str, columns={0 : "file_path"})      

# updating file path column with full path
for i in range(0, len(df1_wav_train)):
    file_name = df1_wav_train['file_path'].iloc[i]
    file_name_full_path = '/home/megha/german dataset/german-speechdata-package-v2/train/'+file_name
    df1_wav_train['file_path'].iloc[i] = file_name_full_path
```

```python
#test


for filename in os.listdir(inputdir_test): 
    if filename.endswith(".xml"):
        #print(filename)     
        df2_test = pd.DataFrame([filename])
        df1_test = df1_test.append(df2_test, ignore_index = True)
    if filename.endswith(".wav"):
        #print(filename)     
        df2_wav_test = pd.DataFrame([filename])
        df1_wav_test = df1_wav_test.append(df2_wav_test, ignore_index = True)
        
df1_test = df1_test.rename(index=str, columns={0 : "xml_files"})
df1_wav_test = df1_wav_test.rename(index=str, columns={0 : "file_path"})

# updating file path column with full path
for i in range(0, len(df1_wav_test)):
    file_name = df1_wav_test['file_path'].iloc[i]
    file_name_full_path = '/home/megha/german dataset/german-speechdata-package-v2/test/'+file_name
    df1_wav_test['file_path'].iloc[i] = file_name_full_path
```
