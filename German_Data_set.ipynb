```python
#1) separate xml file to different dataframes
#df1_dev = pd.DataFrame(columns=['sentence','clean_sentence'])

inputdir_dev = '/home/megha/german dataset/german-speechdata-package-v2/dev'    # this is only dev 
inputdir_train = '/home/megha/german dataset/german-speechdata-package-v2/train' # this is only train 
inputdir_test = '/home/megha/german dataset/german-speechdata-package-v2/test' # this is only test

outputdir = '/home/megha/german dataset/german-speechdata-package-v2'


df1_dev = pd.DataFrame(columns=['cleaned_sentence','sentence'])
df1_train = pd.DataFrame(columns=['cleaned_sentence','sentence'])
df1_test = pd.DataFrame(columns=['cleaned_sentence','sentence'])

df1_wav_dev = pd.DataFrame(columns=['wave_size','transcript'])
df1_wav_train = pd.DataFrame(columns=['wave_size','transcript'])
df1_wav_test = pd.DataFrame(columns=['wave_size','transcript'])

```

```python
#dev


for filename in os.listdir(inputdir_dev): 
    if filename.endswith(".xml"):
        #print(filename)     
        df2_dev = pd.DataFrame([filename])
        df1_dev = df1_dev.append(df2_dev, ignore_index = True)
    if filename.endswith(".wav"):
        #print(filename)     
        df2_wav_dev = pd.DataFrame([filename])
        df1_wav_dev = df1_wav_dev.append(df2_wav_dev, ignore_index = True)
        
df1_dev = df1_dev.rename(index=str, columns={0 : "xml_files"})
df1_wav_dev = df1_wav_dev.rename(index=str, columns={0 : "file_path"})


# updating file path column with full path
for i in range(0, len(df1_wav_dev)):
    file_name = df1_wav_dev['file_path'].iloc[i]
    file_name_full_path = '/home/megha/german dataset/german-speechdata-package-v2/dev/'+file_name
    df1_wav_dev['file_path'].iloc[i] = file_name_full_path
```

```python
#train


for filename in os.listdir(inputdir_train): 
    if filename.endswith(".xml"):
        #print(filename)     
        df2_train = pd.DataFrame([filename])
        df1_train = df1_train.append(df2_train, ignore_index = True)
    if filename.endswith(".wav"):
        #print(filename)     
        df2_wav_train = pd.DataFrame([filename])
        df1_wav_train = df1_wav_train.append(df2_wav_train, ignore_index = True)
        
df1_train = df1_train.rename(index=str, columns={0 : "xml_files"})
df1_wav_train = df1_wav_train.rename(index=str, columns={0 : "file_path"})      

# updating file path column with full path
for i in range(0, len(df1_wav_train)):
    file_name = df1_wav_train['file_path'].iloc[i]
    file_name_full_path = '/home/megha/german dataset/german-speechdata-package-v2/train/'+file_name
    df1_wav_train['file_path'].iloc[i] = file_name_full_path
```

```python
#test


for filename in os.listdir(inputdir_test): 
    if filename.endswith(".xml"):
        #print(filename)     
        df2_test = pd.DataFrame([filename])
        df1_test = df1_test.append(df2_test, ignore_index = True)
    if filename.endswith(".wav"):
        #print(filename)     
        df2_wav_test = pd.DataFrame([filename])
        df1_wav_test = df1_wav_test.append(df2_wav_test, ignore_index = True)
        
df1_test = df1_test.rename(index=str, columns={0 : "xml_files"})
df1_wav_test = df1_wav_test.rename(index=str, columns={0 : "file_path"})

# updating file path column with full path
for i in range(0, len(df1_wav_test)):
    file_name = df1_wav_test['file_path'].iloc[i]
    file_name_full_path = '/home/megha/german dataset/german-speechdata-package-v2/test/'+file_name
    df1_wav_test['file_path'].iloc[i] = file_name_full_path
```

```python
#2) extracting xml elements and updating in the dataframe

for i in range(0, len(df1_dev)):
    file_name = df1_dev['xml_files'].iloc[i]
    xmlDoc = open('/home/megha/german dataset/german-speechdata-package-v2/dev/'+file_name, 'r')
    xmlDocData = xmlDoc.read()
    xmlDocTree = etree.XML(xmlDocData)
    for ingredient in xmlDocTree.iter('sentence'):
        sentence = ingredient.text
        df1_dev['sentence'].iloc[i] = sentence
    for ingredient in xmlDocTree.iter('cleaned_sentence'):
        cleaned_sentence = ingredient.text
        df1_dev['cleaned_sentence'].iloc[i] = cleaned_sentence

        
for i in range(0, len(df1_train)):
    file_name = df1_train['xml_files'].iloc[i]
    xmlDoc = open('/home/megha/german dataset/german-speechdata-package-v2/train/'+file_name, 'r')
    xmlDocData = xmlDoc.read()
    xmlDocTree = etree.XML(xmlDocData)
    for ingredient in xmlDocTree.iter('sentence'):
        sentence = ingredient.text
        df1_train['sentence'].iloc[i] = sentence
    for ingredient in xmlDocTree.iter('cleaned_sentence'):
        cleaned_sentence = ingredient.text
        df1_train['cleaned_sentence'].iloc[i] = cleaned_sentence
        
        
for i in range(0, len(df1_test)):
    file_name = df1_test['xml_files'].iloc[i]
    xmlDoc = open('/home/megha/german dataset/german-speechdata-package-v2/test/'+file_name, 'r')
    xmlDocData = xmlDoc.read()
    xmlDocTree = etree.XML(xmlDocData)
    for ingredient in xmlDocTree.iter('sentence'):
        sentence = ingredient.text
        df1_test['sentence'].iloc[i] = sentence
    for ingredient in xmlDocTree.iter('cleaned_sentence'):
        cleaned_sentence = ingredient.text
        df1_test['cleaned_sentence'].iloc[i] = cleaned_sentence
        
```

```python
#3) extracting wave file size

for i in range(0, len(df1_wav_dev)):
    file_path = df1_wav_dev['file_path'].iloc[i]
    statinfo = os.stat(file_path)
    file_size = statinfo.st_size
    df1_wav_dev['wave_size'].iloc[i] = file_size

    
for i in range(0, len(df1_wav_train)):
    file_path = df1_wav_train['file_path'].iloc[i]
    statinfo = os.stat(file_path)
    file_size = statinfo.st_size
    df1_wav_train['wave_size'].iloc[i] = file_size
    
    
for i in range(0, len(df1_wav_test)):
    file_path = df1_wav_test['file_path'].iloc[i]
    statinfo = os.stat(file_path)
    file_size = statinfo.st_size
    df1_wav_test['wave_size'].iloc[i] = file_size
```

```python
# 4) match the wav files and update the transcript

# dev


for i in range(0, len(df1_wav_dev)): 
    file_path = df1_wav_dev['file_path'].iloc[i]
    fileName = file_path.split('/')[-1]
    fileName_part_1 = fileName.split('_')[0]
    for j in range(0, len(df1_dev)): 
        #print(df1_dev['xml_files'].iloc[j].split('.')[0].replace(" ", ""))
        if( fileName_part_1 == df1_dev['xml_files'].iloc[j].split('.')[0]):
            df1_wav_dev['transcript'].iloc[i] = df1_dev['cleaned_sentence'].iloc[j]
            
    
```

```python
# train


for i in range(0, len(df1_wav_train)): 
    file_path = df1_wav_train['file_path'].iloc[i]
    fileName = file_path.split('/')[-1]
    fileName_part_1 = fileName.split('_')[0]
    for j in range(0, len(df1_train)): 
        #print(df1_dev['xml_files'].iloc[j].split('.')[0].replace(" ", ""))
        if( fileName_part_1 == df1_train['xml_files'].iloc[j].split('.')[0]):
            df1_wav_train['transcript'].iloc[i] = df1_train['cleaned_sentence'].iloc[j]
            
```

```python
# test


for i in range(0, len(df1_wav_test)): 
    file_path = df1_wav_test['file_path'].iloc[i]
    fileName = file_path.split('/')[-1]
    fileName_part_1 = fileName.split('_')[0]
    for j in range(0, len(df1_test)): 
        #print(df1_dev['xml_files'].iloc[j].split('.')[0].replace(" ", ""))
        if( fileName_part_1 == df1_test['xml_files'].iloc[j].split('.')[0]):
            df1_wav_test['transcript'].iloc[i] = df1_test['cleaned_sentence'].iloc[j]
            
```
